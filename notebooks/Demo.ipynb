#### Reading the file in form of DCM format by providing specific path of file.

```python
import pydicom
import matplotlib.pyplot as plt
# Load the DICOM file
ds = pydicom.dcmread('/workspaces/codespaces-jupyter/data/0003.DCM')
# print(ds.Pixel Data)
# Get the pixel array from the DICOM data
pixel_array = ds.pixel_array
```

### By using the shape element, we can observe the number of images are present in single image file.

```python
pixel_array.shape
```

### Plotting the images in a grid format.

```python
plt.figure(figsize=(10,10)) # specifying the overall grid size

for i in range(17):
    plt.subplot(4,5,i+1)    # the number of images in the grid is 4*5 (20)
    plt.imshow(pixel_array[i,:,:])

plt.show()
```
